# Fuel Fuel Optimization: Speed vs. Cost (including wind impact)

**Goal:** Build a simple mathematical model + simulation to estimate fuel consumption and find an "optimal" driving speed under different wind conditions.

**Author:** Alexander

**Date:** 2026-

## 1. Problem Statement

We want to understand how fuel consumption (L/100km) changes with speed and wind.

Then we want to find the speed that minimizes:
- Fuel consumption (L/100km)
- Cost per 100km (EUR/100km)


We will explore multiple wind scenarios:
- no wind
- headwind (wind against the car)
- tailwind (with helping the car)

## 2. Assumptions and Simplifications

This project uses a simplified model. Key assumptions:
1. Road is flat (no elevation gain).
2. Car speed is constant (no acceleration phases).
3. Fuel use has two main components:
   - rolling resistance (roughly increases with speed)
   - aerodynamic drag (roughly increases with speed in terms of fuel per distance)
4. Wind affects the effective air speed, where headwind is positive, tailwind is negative.
5. We ignore traffic, stops, temperature, tire pressure, engine efficiency curves, etc.

The goal is to create a model that is useful and explainable


### More about effective air speed

We define the effective air speed as:

$$ \Large v_{\text{air}} = v + v_{\text{wind}} $$

where:
- $v_{\text{air}} \rightarrow$ effective speed relative to air
- $v \rightarrow$ car speed relative to the ground (dashboard speed)
- $v_{\text{wind}} \rightarrow$ wind speed

Sign convention:
- $v_{\text{wind}} > 0 \rightarrow$ headwind
- $v_{\text{wind}} < 0 \rightarrow$ tailwind
